```python
import matplotlib.pyplot as plt
import numpy as np

# Define number of dipoles
n_values = np.arange(1, 12)

# Compute total matrix elements: 2^n x 2^n = 2^(2n)
matrix_sizes = 2 ** (2 * n_values)

# Compute individual matrix dimensions: 2^n
dimensions = 2 ** n_values

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(n_values, matrix_sizes, marker='o', color='black', linewidth=2)

# Annotate each point with matrix dimensions and total elements
for n, dim, size in zip(n_values, dimensions, matrix_sizes):
    plt.text(n-.2, size + 1e5, f"{dim}×{dim}\n({size:,})",
             fontsize=11, ha='center', va='bottom')

# Set y-limits with space for annotations
plt.ylim(0, matrix_sizes[-1] * 1.25)

# Labels and title
plt.title("Exponential Growth of Tensor Product Operator Size", fontsize=21)
plt.xlabel("Number of Dipoles (n)", fontsize=18)
plt.ylabel("Total Matrix Elements (2ⁿ × 2ⁿ)", fontsize=18)

# Tick labels
plt.xticks(n_values, fontsize=12)
plt.yticks(fontsize=12)

# Grid and layout
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()

# Save and show
plt.savefig("figure4.png", dpi=300)
plt.show()

```
